```python
import pandas as pd

df = pd.read_csv("meningitis_dataset.csv")

df.head(10)
```

```python
# female 1, male 0
# disease column

female = df[df['gender']=='Female'].index
df.loc[female,'gender_index']=1
male = df[df['gender']=='Male'].index
df.loc[male,'gender_index']=0


```

```python
pd.set_option('display.max_columns', None)

diseases = df['disease'].unique()

for disease in diseases :
    filtered = df[df['disease']==disease].index
    df.loc[filtered,disease]=1

    filtered = df[df['disease']!=disease].index
    df.loc[filtered,disease]=0

df.head(10)

```
