```python
sales = pd.read_json('grocery_product_sales.json')
sales.head(10)
```

```python
sales = sales.dropna()
#sales.isna()
#sales.dtypes

```

```python
sales['products'] = sales['products'].astype(str)
sales['products'] = sales['products'].str.split(r'[,|;]')
sales = sales.explode('products', ignore_index=True)

sales['products'] = sales['products'].astype(str)

sales[['products', 'quantity']] = sales['products'].str.split('(', n=1, expand=True)

sales.head(10)
```

```python
sales['quantity'] = sales['quantity'].astype(str)
sales['quantity'] = sales['quantity'].str.extract(r'(\d)')
sales['quantity'] = sales['quantity'].astype(int)

sales.head(10)
```
