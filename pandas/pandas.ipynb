```python
iv = pd.read_json("dim_customer.json")
df = pd.read_csv("fct_invoice.csv")



iv.rename(columns={"id":"customer_id"})

#iv.head(5)
# df.head(5)
# 1. Demographics related to shopping mall : How many males and females are in each shopping mall, average age

merged = pd.merge(df,iv,left_on="customer_id",right_on="id")
merged.sort_values(by=['shopping_mall'])

```
